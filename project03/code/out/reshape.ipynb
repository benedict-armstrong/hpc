```python
import pandas as pd
import json
```

```python
with open("strong_scaling_results.json") as f:
    weak = json.load(f)

with open("strong_scaling_results.json") as f:
    strong = json.load(f)
```

```python
weak
```

```python
weak_df = pd.DataFrame()

times = weak["times"]
threads = weak["threads"]
sizes = weak["sizes"]
runs = weak["runs"]

for i, t in enumerate(times):
    s = sizes[i]
    thread = threads[i]
    for j, time in enumerate(t):
        temp_df = pd.DataFrame({
            "cpus": thread,
            "res": s,
            "base_res": s,
            "time": time,
            "run": j + 1,
        })

        weak_df = pd.concat([weak_df, temp_df])


# save to csv
weak_df.to_csv(
    "../../../project05/code/mini_app/out/results_weak_omp.csv", index=False)
weak_df
```

```python
# repeat for strong scaling
strong_df = pd.DataFrame()

times = strong["times"]
threads = strong["threads"]
sizes = strong["sizes"]
runs = strong["runs"]

for i, t in enumerate(times):
    s = sizes[i]
    thread = threads[i]
    for j, time in enumerate(t):
        temp_df = pd.DataFrame({
            "cpus": thread,
            "res": s,
            "time": time,
            "run": j + 1
        })

        strong_df = pd.concat([strong_df, temp_df])


# save to csv
strong_df.to_csv(
    "../../../project05/code/mini_app/out/results_strong_omp.csv", index=False)
strong_df
```
