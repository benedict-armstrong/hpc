```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
lines = []
with open('perf.data') as f:
    lines = f.readlines()

# remove all empty lines and strip leading/trailing whitespace
lines = [line.strip().split() for line in lines if line.strip()]
```

```python
# group by number of processes
grouped = {}
for line in lines:
    n = int(line[0])
    if n not in grouped:
        grouped[n] = []
    grouped[n].append(float(line[2]))

# grouped
```

```python
# plot bar chart x-axis is mpi_rank and y-axis is time, offset by 0.2 for each n
fig, ax = plt.subplots(figsize=(10, 5))
ax.grid()
for i, (n, times) in enumerate(grouped.items()):
    width = 0.5
    overlap = 0.2
    offset = -width / 2 + i * overlap
    ax.bar(np.arange(len(times)) + offset + 1, times, width=width, label=f'{n} processes')


ax.set_xticks(np.arange(len(times)) + 1)
ax.set_xlabel('MPI Rank')
ax.set_ylabel('Time (s)')
ax.set_title('Performance of MPI Mandelbrot set computation')
ax.legend()
```
